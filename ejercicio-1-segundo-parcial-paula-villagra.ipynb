```julia
#Paula Jazmin Villagra Zayas
# 1er. ejercicio 2do. examen parcial

#Problema primal
using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin 
        0 <= x1             # Nro. de hectareas dedicadas al cultivo de trigo
        0 <= x2             # Nro. de hectareas dedicadas al cultivo de avena
        0 <= x3             # Nro. de hectareas dedicadas al cultivo de centeno
end)  

@objective(m, Max, 60*x1 + 100*x2 + 80*x3)     # Beneficio maximo para el agricultor

@constraint(m, M1, 6*x1 + 8*x2 + 10*x3 <= 5000)           # Cantidad de horas hombre disponibles (M1)     
@constraint(m, M2, 100*x1 + 150*x2 + 120*x3 <= 60000)     # Cantidad de dinero disponible (M2)
@constraint(m, M3, x1 + x2 + x3 <= 500)                   # Cantidad de hectareas disponibles (M3)
print(m)

JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x1_trigo=JuMP.value(x1)
x2_avena=JuMP.value(x2)
x3_centeno=JuMP.value(x3)

println("Beneficio maximo obtenido =", obj_value)
println("Cantidad de hectareas para cultivo de trigo =", x1_trigo)
println("Cantidad de hectareas para cultivo de avena =", x2_avena)
println("Cantidad de hectareas para cultivo de centeno =", x3_centeno)
```

```julia
#Si el agricultor contrata 500 horas-hombre más, por 1500 euros, su beneficio sería
using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin 
        0 <= x1             # Nro. de hectareas dedicadas al cultivo de trigo
        0 <= x2             # Nro. de hectareas dedicadas al cultivo de avena
        0 <= x3             # Nro. de hectareas dedicadas al cultivo de centeno
end)  

@objective(m, Max, 60*x1 + 100*x2 + 80*x3)     # Beneficio maximo para el agricultor

@constraint(m, M1, 6*x1 + 8*x2 + 10*x3 <= 5500)           # Cantidad de horas hombre disponibles (M1)     
@constraint(m, M2, 100*x1 + 150*x2 + 120*x3 <= 58500)     # Cantidad de dinero disponible (M2)
@constraint(m, M3, x1 + x2 + x3 <= 500)                   # Cantidad de hectareas disponibles (M3)
print(m)

JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x1_trigo=JuMP.value(x1)
x2_avena=JuMP.value(x2)
x3_centeno=JuMP.value(x3)

println("Beneficio maximo obtenido con 5500 horas-hombre y 58500 euros =", obj_value)
println("Cantidad de hectareas para cultivo de trigo =", x1_trigo)
println("Cantidad de hectareas para cultivo de avena =", x2_avena)
println("Cantidad de hectareas para cultivo de centeno =", x3_centeno)
```

```julia
#Si se cultivan al menos 100 hectareas de trigo, el beneficio por ha de trigo es de 60 euros y el beneficio maximo es:

using JuMP, Clp

m = Model(with_optimizer(Clp.Optimizer))

@variables(m, begin 
        0 <= x1             # Nro. de hectareas dedicadas al cultivo de trigo
        0 <= x2             # Nro. de hectareas dedicadas al cultivo de avena
        0 <= x3             # Nro. de hectareas dedicadas al cultivo de centeno
end)  

@objective(m, Max, 60*x1 + 100*x2 + 80*x3)     # Beneficio maximo para el agricultor

@constraint(m, M1, 6*x1 + 8*x2 + 10*x3 <= 5000)           # Cantidad de horas hombre disponibles (M1)     
@constraint(m, M2, 100*x1 + 150*x2 + 120*x3 <= 60000)     # Cantidad de dinero disponible (M2)
@constraint(m, M3, x1 + x2 + x3 <= 500)                   # Cantidad de hectareas disponibles (M3)
@constraint(m, M4, x1  >= 100)                   # Cantidad de hectareas minimas de trigo (M4)
print(m)

JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x1_trigo=JuMP.value(x1)
x2_avena=JuMP.value(x2)
x3_centeno=JuMP.value(x3)

println("Beneficio maximo obtenido =", obj_value)
println("Cantidad de hectareas para cultivo de trigo =", x1_trigo)
println("Cantidad de hectareas para cultivo de avena =", x2_avena)
println("Cantidad de hectareas para cultivo de centeno =", x3_centeno)
```
